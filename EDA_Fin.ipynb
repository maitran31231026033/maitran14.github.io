```python
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
import datetime as dt
%matplotlib inline

# sửa lỗi tải dữ liệu của pandas_datareader
from IPython.display import display
import yfinance as yf
# Removed: yf.pdr_override()
```

```python
end = dt.datetime.now()
start = dt.datetime(2016, 1, 1)
print(f'{start} -> {end}')

# Fetch data using yfinance
df = yf.download('AAPL', start=start, end=end)
```

```python
df['Close'].plot(legend=True, figsize=(12,8), title='AAPL', label = 'Close')
plt.show()
```

```python
# Flatten the column MultiIndex manually by creating a new list of column names
new_columns = []
for col_tuple in df.columns:
    if 'Price' in col_tuple:
        new_columns.append(col_tuple[1])
    elif 'Volume' in col_tuple:
        new_columns.append(col_tuple[0])

df.columns = new_columns


# Now plot using mplfinance
mpf.plot(df, type='line', style='yahoo', volume=True, figscale=1.5, title='AAPL Price and Volume')
plt.show()
```
